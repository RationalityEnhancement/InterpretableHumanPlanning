```python
from scipy.special import logsumexp
from scipy.stats import multivariate_normal
import math
def compute_error_grads(data, W, K, F):
    N = len(data)
    W = W[:-K].reshape(-1, F)
    pi = W[-K:]
    negative_data_grads = np.zeros((N, K, F))
    negative_log_probs = np.zeros((N, K))
    
    for n in range(N):
        for k in range(K):
            negative_log_probs[n][k], negative_data_grads[n][k] = data[n].compute_error_grad(W[k])
    
    return negative_log_probs, negative_data_grads

def f(weights, total_data, num_clusters, num_features):
    prior = multivariate_normal(mean = np.zeros(num_features), cov=np.eye(num_features))
    negative_log_probs, negative_data_grads  = compute_error_grads(total_data, weights, num_clusters, num_features)
    W = weights[:-num_clusters].reshape(-1, num_features)
    pi = weights[-num_clusters:]
    log_pi = np.log(pi)
    total_log_probs = log_pi - negative_log_probs
    E = -np.sum(logsumexp(total_log_probs, axis=1))
    #E -= np.sum([prior.logpdf(W[k]) for k in range(num_clusters)])
    print(E)
    total_grads = np.zeros((num_clusters, num_features))
    for n in range(len(total_data)):
        D = np.exp(logsumexp(total_log_probs[n]))
        for k in range(num_clusters):
            total_grads[k] += (pi[k]*negative_data_grads[n][k]*np.exp(-negative_log_probs[n][k]))/D
#     for k in range(num_clusters):
#         total_grads[k] += W[k]
    grad_pi = np.zeros_like(pi)
    for n in range(len(total_data)):
        D = logsumexp(total_log_probs[n])
        for k in range(num_clusters):
            grad_pi[k] += np.exp(-negative_log_probs[n][k] - D)
    E_grad = np.append(total_grads.flatten(), grad_pi)
    #print(E, E_grad)
    return E, E_grad
```

```python
def summation_constraint(weights, num_clusters):
    res = np.sum(weights[-num_clusters:]) - 1
    return res

def positive_constraint(weights, k, num_clusters):
    return weights[-num_clusters+k:] - 1e-4
```

```python
from scipy.optimize import minimize
W = np.random.randn(num_clusters, num_features)
pi = np.ones(num_clusters)/num_clusters
weights = np.append(W.flatten(), pi)
print(weights.shape)
cons = [{'type': 'eq', 'fun': lambda x: summation_constraint(x, num_clusters)}] + [{'type': 'ineq', 'fun': lambda x: positive_constraint(x, num_clusters, k)} for k in range(num_clusters)]
res = minimize(lambda x: f(x, total_data, num_clusters, num_features), weights, jac = True, constraints=cons)
```

```python
def get_experiment_strategies(exp_num, block=None):
    path = "results/final_strategy_inferences"
    if block:
        strategies = pickle_load(f"{path}/{exp_num}_{block}_strategies.pkl")
    else:
        strategies = pickle_load(f"{path}/{exp_num}_strategies.pkl")
    return strategies

def print_clusters(cluster_info):
    for cluster in sorted(list(cluster_info.keys())):
        print("\nCluster Number", cluster, "\n")
        for p in cluster_info[cluster]:
            print(p)

def get_ds_weights(weights, features, feature_systems, decision_systems):
    ds_weights = defaultdict(int)
    for (f, w) in zip(features, weights):
        if feature_systems[f] == 'stopping_criteria' or f == "soft_pruning" or f == "termination_constant" or f=="is_previous_successor_negative":
            ds_weights[feature_systems[f]] += -w
        else:
            ds_weights[feature_systems[f]] += w
    return np.array([ds_weights[ds] for ds in decision_systems])

def get_W_DS(W, feature_systems, decision_systems):
    num_strategies = W.shape[0]
    W_DS = []
    for i in range(num_strategies):
        W_DS.append(get_ds_weights(W[i], features, feature_systems, decision_systems))
    W_DS = np.array(W_DS)
    return W_DS

def get_ds_proportions(W_DS):
    DS_proportions = np.zeros_like(W_DS)
    for i in range(DS_proportions.shape[0]):
        total_w = np.sum([abs(w) for w in W_DS[i]])
        if total_w != 0:
            DS_proportions[i] = abs(W_DS[i])/total_w
    return DS_proportions
```
